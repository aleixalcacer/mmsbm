```python
import psycopg2
import pandas as pd
import numpy as np
```

```python
from queries import *
from secret import *
```

```python
classroom_id = 76
```

```python
conn = psycopg2.connect(
    host=host,
    database=database,
    user=user,
    password=password)
cur = conn.cursor()
```

```python
# Cognitive
cur.execute(get_cog(classroom_id))
cog = pd.DataFrame(cur.fetchall())
cog.columns = ["name", "id", "session", "item", "time", "correct", "incorrect"]
```

```python
# Adaptive
cur.execute(get_ada(classroom_id))
ada = pd.DataFrame(cur.fetchall())
ada.columns = ["detail", "id", "session", "result_id", "session_activity", "item", "item_type"]
```

```python
# Sociogram
cur.execute(get_soc(classroom_id))
soc = pd.DataFrame(cur.fetchall())
soc.columns = ["time", "result_id", "other_student", "action", "id", "session", "item", "session_activity"]
```

```python
# Gender
cur.execute(get_gen(classroom_id))
gen = pd.DataFrame(cur.fetchall())
gen.columns = ["select_id", "results_activity", "detail", "id", "session", "action", "item", "session_activity"]
gen["gender"] = np.where(gen["detail"] == "48977", "girl", "boy")
```

```python
# list of activity items
cur.execute(get_act_items())
actitems = pd.DataFrame(cur.fetchall())
actitems.columns = ["activity_id", "description", "type", "item"]
```

```python
# Give the data the activities they belong to
cog = cog.merge(actitems, on="item")
ada = ada.merge(actitems, on="item")
soc = soc.merge(actitems, on="item")
```
